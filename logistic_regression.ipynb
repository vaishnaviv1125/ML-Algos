```python
#import the required libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#load the dataset

card_data=pd.read_csv("https://storage.googleapis.com/download.tensorflow.org/data/creditcard.csv")
```

```python
#display the 1st 5 rows of the dataset
card_data.head()
```

```python
#display tall the rows and columns of the data
card_data.shape
```

```python
#check the missing values
card_data.isnull().sum()        #checks each cell in the DataFrame and returns True (1) if the value is missing , otherwise False (0)
```

```python
#detail info of dataset
card_data.info()
```

```python
#label distribution of class  #counts the number of instances for each class
card_data['Class'].value_counts()
```

```python
#separate the legit and fraud trans
legit=card_data[card_data.Class==0]
fraud=card_data[card_data.Class==1]
```

```python
#statistical measures of both
print(legit.Amount.describe())
print(fraud.Amount.describe())
```

```python
#compare the trans
card_data.groupby("Class").mean()
```

```python
#legit is class 1
#fraud is class 0
```

```python
#Undersampling


legit_sample=legit.sample(n=492)
```

```python
new_data=pd.concat([legit_sample,fraud],axis=0) #
```

```python
new_data.shape #displays all the rows and columns
```

```python
new_data.head() #displays the 1st 5 rows
```

```python
new_data.tail()   #displays the last 5 rows
```

```python
new_data.isnull().sum()
```

```python
new_data['Class'].value_counts() #undersampling is done (both the calsses have equal data)
```

```python
card_data.groupby("Class").mean()
```

```python
X=new_data.drop(columns="Class",axis=1)
Y=new_data["Class"]
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print('total data:',X.shape)
print('train data:',X_train.shape)
print('test_data:',X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("Accuracy on training data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy on training data: ", test_data_accuracy)
```

```python
import numpy as np
import gradio as gr

# Assuming model is your trained credit card fraud detection model

def detect_fraud(*features):
    try:
        # Convert all input features to a numpy array
        input_data = np.array(features, dtype=float)

        # Reshape the data for a single prediction
        input_data_reshaped = input_data.reshape(1, -1)

        # Make prediction using the loaded model
        prediction = model.predict(input_data_reshaped)

        # Interpret the prediction
        if prediction[0] == 0:
            return "It's a Legit transaction"
        else:
            return "It's a Fraud transaction"
    except Exception as e:
        return str(e)

# Define Gradio interface with 10 numeric inputs
inputs = [gr.Number(label=f"Feature {i+1}") for i in range(10)]

demo = gr.Interface(
    fn=detect_fraud,
    inputs=inputs,
    outputs="text",
    title="Credit Card Fraud Detection",
    description="Enter the 10 transaction features to predict if it's legitimate or fraudulent."
)

# Run on localhost
demo.launch()
```
